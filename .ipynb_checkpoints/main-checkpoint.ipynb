# Как решить задачу

Пусть существует некая функция, которая решает поставленную задачу.
Или существует закон, описывающий интересное нам явление.
$$F(x)=y$$
Мы по какой-то причине не может этот закон явно выразить так, 
чтобы можно было написать программу (алгоритм), который бы .
Или можем, но результат нас не устраивает (слишком сложно явно описать все ситуации).
Или у нас если алгоритм, но он слишком медленный.
...
Тогда наша задача найти **аппроксимацию** (приближение) этого закона.
Искать приближение мы будем на основе данных (информации, наблюдений).

```python
from matplotlib import pyplot as plt
from scipy import interpolate
from scipy.optimize import curve_fit
import numpy as np

def func(x, a, b, c):
    return a*(x**3) + b*(x**2) + c

dX = np.arange(0,9,0.01)
X = list(range(10))
Y = [1, 4, 5, 4.5, 6.5, 6, 5, 5.5, 8, 10]

f_int = interpolate.interp1d(X, Y, kind='cubic')
Y_int = f_int(dX)

popt, pcov = curve_fit(func, X, Y)

fig = plt.figure(figsize=(6,6))
plt.plot(X, Y, 'b.', markersize=14, label='измерения')
plt.plot(dX, Y_int, 'g-', label='интерполяция')
plt.plot(dX, func(dX, *popt), 'r-', label='аппроксимация')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})
plt.show()
```

Наше приближение это:
$$f_{\theta}(x)=\tilde{y}$$

Мы хотим, чтобы $F(x) \simeq f_{\theta}(x)$ на некотором множестве $X$. 

Назовем процесс поиска такого $f_{\theta}$ на основе некоторых известных данных $\tilde{X}$ - **обучением**, а сами данные ($\tilde{X} \subset X$, выборка из основого множества $X$) - **датасетом**.

Термины:
- $f_{\theta}$ модель;
- $\theta$ параметры модели;
- $x$ произвольный элемент множества $X$;
- $y$ известное (в некотором смысле) истинное значение для элемента $x$;
- $\tilde{y}$ модельное (результат применения модели) значение для элемента $x$.

Почему мы хотим найти именно *аппроксимацию*, а не *интерполяцию*?

Интерполяция в некотором смысле - это *запоминание* датасета, а датасет - это всегда подмножество (*выборка*) несоизмеримо малое из основного множества.
Например, если мы хотим построить на основе данных систему, способную на произвольной фотографии находить участки с собаками, то наше основное множество - это фотографии всех собак - живших, существующих и еще не родившихся, снятные со всех возможных ракурсах и т.д.  При этом скорее всего еще потребуется **разметить** каждую фотографию из этого бесконечного множества (некоторомы образом задать $y$).

Вместо того, чтобы на 100% ориентироваться на одно только запоминание, мы хотим, чтобы наша модель была способна *обобщаться* и корректно работать на новых данных. 

Иначе говоря, модель должна *описывать* некоторую информационную связь внутри множества данных.

Рассмотрим простой пример. Зададим линейный закон $y(x) = ax + b$, предположим, что в процессе измерения возникает ошибка (шум) $\xi$, например, *аддитивный* $\widetilde{y}=y+\xi$. Измерения проводились с некоторым случайным переменным шага, тогда набор данных $D$ представляет собой набор пар $D=\{(x_i,\widetilde{y}_i), i=1..n\}$, где $n$ - число измерений (число элементов датасета, число сэмплов, размер выборки и т.д.).

График датасета (облака точек) для $y(x) = 4x + 2$, сгенерированный подобным образом, представлен ниже.

```python
real_a = 4
real_b = 2

X = 10*np.random.rand(100,1)
Y = real_a*X + real_b + np.random.randn(100,1)*6

dX = np.linspace(0,10,100)
Y_gt = real_a*dX + real_b

plt.plot(dX, Y_gt, 'g-', linewidth=3, label='y(x)=4x+2')
plt.plot(X, Y, 'r.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})
plt.show()
```

Зеленая прямая - это истинный закон, который в реальной ситуации нам неизвестен, и мы хотим его найти (построить модель) на основе данных.
Пусть нам известна только часть датасета, например при $x<6$, закрасим "неизвестные" точки в синий цвет и попробуем построить гладкую интерполяцию кубическим сплайном (https://ru.wikipedia.org/wiki/%D0%9A%D1%83%D0%B1%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B9_%D1%81%D0%BF%D0%BB%D0%B0%D0%B9%D0%BD) по всем "известным" точкам (красный цвет). Оценим как будет себя вести полученный закон на новых "неизвестных" точках (экстраполяция).

```python
dX_inter = np.sort(np.concatenate([dX, X[:,0]],-1))

f_int_linear = interpolate.interp1d(
    X[X<6], Y[X<6], kind='cubic', 
    assume_sorted=False, bounds_error=False, fill_value='extrapolate')
Y_int_linear = f_int_linear(dX_inter)

fig = plt.figure(figsize=(14,4))

fig.add_subplot(1,2,1)
plt.plot(dX_inter, Y_int_linear, 'm-', label='сплайн')
plt.plot(dX, Y_gt, 'g-', linewidth=3, label='y(x)=4x+2')
plt.plot(X[X<6], Y[X<6], 'r.', markersize=8)
plt.plot(X[X>=6], Y[X>=6], 'b.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

fig.add_subplot(1,2,2)
plt.plot(dX_inter, Y_int_linear, 'm-', label='сплайн')
plt.plot(dX, Y_gt, 'g-', linewidth=3, label='y(x)=4x+2')
plt.plot(X[X<6], Y[X<6], 'r.', markersize=8)
plt.plot(X[X>=6], Y[X>=6], 'b.', markersize=8)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

plt.show()
```

Видно, что модель идеально проходит через красные точки, при этом на новых данных модель не попадает ни в одну точки и порядок ошибки составляет $10^ {-6}$, следовательно модель непригодна для задачи экстраполяции, прогнозирования. Посмотрим, что будет происходить, если новые точки будут находится в одном интервале по $x$ вместе с известными точками.

```python
dX_inter = np.sort(np.concatenate([dX, X[:,0]],-1))

f_int_linear = interpolate.interp1d(
    X[:50,0], Y[:50,0], kind='cubic',
    assume_sorted=False, bounds_error=False, fill_value='extrapolate')
Y_int_linear = f_int_linear(dX_inter)


fig = plt.figure(figsize=(14,4))

fig.add_subplot(1,2,1)
plt.plot(dX_inter, Y_int_linear, 'm-', label='сплайн')
plt.plot(dX, Y_gt, 'g-', linewidth=3, label='y(x)=4x+2')
plt.plot(X[:50], Y[:50], 'r.', markersize=8)
plt.plot(X[50:], Y[50:], 'b.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

fig.add_subplot(1,2,2)
plt.plot(dX_inter, Y_int_linear, 'm-', label='сплайн')
plt.plot(dX, Y_gt, 'g-', linewidth=3, label='y(x)=4x+2')
plt.plot(X[:50], Y[:50], 'r.', markersize=8)
plt.plot(X[50:], Y[50:], 'b.', markersize=8)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

plt.show()
```

Кажется, что ситуация не настолько плохая, как в случае с экстраполяцией, модель проходит ближе к новым неизвестным точкам, хотя на графике с большим масштабом заметно, что местами ошибка в десятки раз превосходит масштаб измерений. Попробуем теперь проделать то же самое только при условии, что известных нам точек значительно меньше новых.

```python
dX_inter = np.sort(np.concatenate([dX, X[:,0]],-1))

f_int_linear = interpolate.interp1d(
    X[:10,0], Y[:10,0], kind='cubic',
    assume_sorted=False, bounds_error=False, fill_value='extrapolate')
Y_int_linear = f_int_linear(dX_inter)


fig = plt.figure(figsize=(14,4))

fig.add_subplot(1,2,1)
plt.plot(dX_inter, Y_int_linear, 'm-', label='сплайн')
plt.plot(dX, Y_gt, 'g-', linewidth=3, label='y(x)=4x+2')
plt.plot(X[:10], Y[:10], 'r.', markersize=8)
plt.plot(X[10:], Y[10:], 'b.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

fig.add_subplot(1,2,2)
plt.plot(dX_inter, Y_int_linear, 'm-', label='сплайн')
plt.plot(dX, Y_gt, 'g-', linewidth=3, label='y(x)=4x+2')
plt.plot(X[:10], Y[:10], 'r.', markersize=8)
plt.plot(X[10:], Y[10:], 'b.', markersize=8)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

plt.show()
```

По графикам видно, что если бы мы искали аппроксимацию, и смогли бы оценить параметры "неизвестной" прямой $y(x)$, то ошибка была бы значительно меньше, чем в случае интерполяции сплайном.

Все рассмотренные ситуации в терминах машинного обучения называют **переобучением** (**overfitting**).

Важно отметить, что *шум* в измерениях может быть любой природы, и с точки зрения теории информации, мы можем называть шумом любую информацию, которая в рамках некоторой задачи нам не нужна или мешает. Например, если мы решаем задачу опредления есть ли собака на произвольной фотографии, нам совершенно не интересно в какой точке земного шара была сделана фотография, где собака находится (если она есть), с какого ракурса она была получена, на какую камеру, при каком освещении, нам даже не интересно какой она породы и размера, и т.д. Все эти факторы вносят свой вклад в то, как будет выглядеть конечная фотография, но для нашей задачи - это все можно считать шумом, и мы хотим, чтобы наша итоговая модель была к нему *инварианта*.

Поэтому независимо от того для какой задачи мы хотим построить модель, датасет всегда будет **зашумлен**, при этом для нелинейного многомерного множества с нелинейным вхождением шума неопределенной (с точки зрения теории вероятности) природы становится очевидно невозможность построение обычной интерполяции по известным точкам для получения *адекватной* модели.

## Как построить модель

Естественным образом, задаемся вопросами:
1. Как определить вид модели $f_{\theta}$?
2. Как найти параметры модели ${\theta}$?
3. Как оценить работу модели?

В качестве модели возьмем многослойный перцептрон (МСП / multilayer perceptron / MLP).
Почему? Известно следующее:
- МСП является **универсальным аппроксиматором**. То есть может *аппроксимировать* любую непрерывную функцию многих переменных с любой точностью. https://en.wikipedia.org/wiki/Universal_approximation_theorem https://ru.wikipedia.org/wiki/%D0%A2%D0%B5%D0%BE%D1%80%D0%B5%D0%BC%D0%B0_%D0%A6%D1%8B%D0%B1%D0%B5%D0%BD%D0%BA%D0%BE
- МСП самая *простая* модель глубокого обучения;
- Перцептрон впервые появился в работах *Розенблатта* https://ru.wikipedia.org/wiki/%D0%9F%D0%B5%D1%80%D1%86%D0%B5%D0%BF%D1%82%D1%80%D0%BE%D0%BD. Модель перцептрона вдохновлена работой нервной системы у высших форм жизни, считается первой искусственной нейронной сетью (математической моделью).

Например, 3-х слойный МСП (5 элементов на 1-ом слое, 3 -- на 2-ом, 1 -- на 3-ем) схематично можно изобрать так:

![MLP](./mlp.png "MLP")

Считаем, что первый слой принимает входной сигнал (входные данные, вектор). Дальше, каждый узел (искусственный нейрон)
принимает значение в зависимости от значений всех нейронов на слое выше с учетом весов-связи. То есть каждый нейрон
в перцептроне (кроме первого входного слоя) связан со всему нейронами на предыдущих слоях. 

Обозначим значения нейронов так: $\{(x_{11}, x_{12}, x_{13}, x_{14}, x_{15}), (x_{21}, x_{22}, x_{23}), (x_{31})\}$

Входные значения на первом слое: $(x_{11}, x_{12}, x_{13}, x_{14}, x_{15})$.
Выходное значение $x_{31} = \tilde{y}$.

Тогда для значение произвольного элемента $x_{ij}$ (кроме 1-ого слоя, он входной) можно вычислить так:
$$
x_{ij} = f\bigg(\sum_{k \in {\text{все нейроны выше}}} x_{i-1,k} {\theta^{i-1}_{j,k}}\bigg)
$$

Это эквивалентно матричному произведению:
$$
x_{i} = f\bigg(\Theta^{T}_{i-1} x_{i-1}\bigg)
$$

То есть считаем, что
- входной вектор - это $x_{0}$;
- $x_{i}$ - вектор на $i$-ом слое;
- $\Theta_i$ - матрица весов для $i$-ого слоя;
- $f()$ - некоторая **нелинейная** функция, называемая ф-ей **активации**, которая применяется к каждому элементу вектора.

Для большей выразительности сети, обычно, считают, что в каждом слое (кроме выходного) есть так называемый нейрон смещения (bias), который всегда равен $1$.

![MLP + bias](./mlp_bias.png "MLP + bias")

В качестве примера рассмотрим двухслойный перцептрон (5,3) c линейной функций $f(x)=x$ (иногда называемой ф-ей идентичности, *identity function*).

$$\{(x_{11}, x_{12}, x_{13}, x_{14}, x_{15}), (x_{21}, x_{22}, x_{23})\}$$

Нейроны $x_{11}..x_{15}$ -- входные нейроны, тогда значения нейронов на 2-ом (выходном) слое:

$$
\begin{align}
x_2 &= \begin{bmatrix}
x_{21} \\ x_{22} \\ x_{23}
\end{bmatrix} &= 
f\bigg(\Theta^{T} x_{1}\bigg) &=
\Theta^{T} x_{1} &=
\begin{bmatrix}
\theta^{1}_{11} \theta^{1}_{12} \theta^{1}_{13} \theta^{1}_{14} \theta^{1}_{15} \theta^{1}_{16} \\
\theta^{1}_{21} \theta^{1}_{22} \theta^{1}_{23} \theta^{1}_{24} \theta^{1}_{25} \theta^{1}_{26} \\
\theta^{1}_{31} \theta^{1}_{32} \theta^{1}_{33} \theta^{1}_{34} \theta^{1}_{35} \theta^{1}_{36}
\end{bmatrix}
\cdot \begin{bmatrix}
x_{11} \\ x_{12} \\ x_{13} \\ x_{14} \\ x_{15} \\ 1
\end{bmatrix} &=
\begin{bmatrix}
\theta^{1}_{11}x_{11} + \theta^{1}_{12}x_{12} + \theta^{1}_{13}x_{13} + \theta^{1}_{14}x_{14} + \theta^{1}_{15}x_{15} + \theta^{1}_{16} \\
\theta^{1}_{21}x_{11} + \theta^{1}_{22}x_{12} + \theta^{1}_{23}x_{13} + \theta^{1}_{24}x_{14} + \theta^{1}_{25}x_{15} + \theta^{1}_{26} \\
\theta^{1}_{31}x_{11} + \theta^{1}_{32}x_{12} + \theta^{1}_{33}x_{13} + \theta^{1}_{34}x_{14} + \theta^{1}_{35}x_{15} + \theta^{1}_{36}
\end{bmatrix}
\end{align}
$$

То есть такая модель содержит 18 параметров (весов) и линейно отображает вектора длиной $5$ в вектора длиной $3$ ($\mathbb{R^5 \rightarrow R^3}$).

Важно отметить, что в МО поиск параметров для некоторого линейного отображения $\mathbb{R^n \rightarrow R^m}$ называется **линейной регрессией**.
Эту задачу полезно рассмотреть отдельно, чтобы найти ответ на вопрос *как найти параметры модели ${\theta}$*.

### Нормальное уравнение

Если в качестве функции ошибки выбрать *MSE* (Mean Squared Error, среднеквадратичная ошибка) $\mathcal{L}(\Theta; h_{\theta}) = \frac{1}{n} \sum (\hat{y} - y)^2$, то выразить оптимальные параметры модели (минимизирующие функцию ошибки) можно аналитически (или в закрытой форме) (*closed form solution*).

$$
\hat{\theta} = \big( X^{T}X \big)^{-1} X^{T} y
$$

#### Правила матричного дифференцирования

Рассмотрим правила матричного дифференцирования, которые будем использовать позднее:
1.   Пусть a и b -- вектора-строки размерности k и U -- симметричная матрица порядка k, тогда $\frac{d}{db}[a^Tb]=\frac{d}{db}[b^Ta]=a.$
2.   Пусть a и b -- вектора-строки размерности k и U -- симметричная матрица порядка k, тогда $\frac{d}{db}[b^TUb]=2Ub.$

#### Вывод формулы

$$
\begin{align}
\mathcal{L}(\Theta; h_{\theta}) = \frac{1}{n} \sum_{i=1}^{n} (\hat{y} - y)^2=\frac{1}{n} \sum_{i=1}^{n} (h_\theta(x^{(i)}) - y)^2
\end{align}
$$
Элемент суммы, а именно $h_\theta(x^{(i)}) - y$, можно представить в виде $\theta^T*x^{(i)}-y^{(i)}$. В векторе-столбце $X*\theta-y$ $i$-й элемент имеет вид $\theta^T*x^{(i)}-y^{(i)}$. Для того, чтобы получить сумму квадратов всех элементов вектора, необходимо умножить вектор на транспонированный вектор:
$$
\begin{align}
\mathcal{L}(\Theta; h_{\theta})=(X*\theta-y)^T*(X*\theta-y)
\end{align}
$$ 

Продифференцируем функцию:

$$
\begin{align}
\frac{d\mathcal{L}(\Theta; h_{\theta})}{d\theta}=\frac{d[(X*\theta-y)^T*(X*\theta-y)]}{d\theta}=\frac{d[((X*\theta)^T-y^T)*(X*\theta-y)]}{d\theta}=\frac{d[(\theta^T*X^T-y^T)*(X*\theta-y)]}{d\theta}=\frac{d[\theta^TX^TX\theta-\theta^TX^Ty-y^TX\theta+y^Ty
]}{d\theta}=
\end{align}
$$

Заметим, что $\theta^TX^Ty$ и $y^TX\theta$ -- матрицы, имеющие размерность 1$\times$1, и $(\theta^TX^Ty)^T=y^TX\theta$, следовательно, $\theta^TX^Ty=y^TX\theta$.

$$
\begin{align}
=\frac{d[\theta^TX^TX\theta]}{d\theta}-2\frac{d[y^TX\theta]}{d\theta}+\frac{d[y^Ty]}{d\theta}=
\end{align}
$$
Применим правила матричного дифференцирования и приравняем производную к 0, это является необходимым условие экстремума функции. Найденная точка экстремума будет минимумом, так как наша функция выпукла, что является достаточным условием минимума.

$$
\begin{align}
=2X^TX\theta-2X^Ty=0
\end{align}
$$

$$
\begin{align}
2X^TX\theta=2X^Ty
\end{align}
$$ 


$$
\begin{align}
(X^TX)^{-1}(X^TX)\theta=(X^TX)^{-1}(X^Ty)
\end{align}
$$ 

$$
\begin{align}
\theta=(X^TX)^{-1}(X^Ty)
\end{align}
$$ 

#### Вывод формулы альтернативный

Дано:
$$
X \theta = y \tag{1}
$$
Так как $X$ не квадратная матрица, не существует обратной $X^{-1}$. При этом $X^T X$ -- квадратная матрица. Умножим уравнение $(1)$ слева на $X^T$:

$$
X^T X \theta = X^T y. \tag{2}
$$

Заметим, что полученное уравнение не эквивалентно исходному.

Теперь, умножив уравнение $(2)$ на $(X^T X)^{-1}$ слева, выразим значения параметра $\theta$:

$$
\theta=(X^TX)^{-1}(X^Ty).
$$

**Замечание:** $(X^TX)^{-1}X^T$ -- называется псевдообратной матрицей Мура–Пенроуза (Moore–Penrose pseudo-inverse).
    
</blockquote>

```python
#найдем вектор theta
X_b = np.c_[np.ones((100, 1)), X] # добавляем x0 = 1 для каждого набора данных
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)# inv () из модуля linear algebra вернет обратную матрицу, метод dot () умножает матрицы
print(theta_best)

X_new = np.array ( [ [ 0 ] , [ 10 ] ] )
X_new_b = np.c_ [ np.ones ( ( 2 , 1 )), X_new]  # добавляем x0 = 1 для каждого набора данных
y_predict = X_new_b.dot( theta_best )

fig = plt.figure(figsize=(14,4))

fig.add_subplot(1,2,1)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(X_new, y_predict, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

fig.add_subplot(1,2,2)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(X_new, y_predict, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.xlim(2,4)
plt.ylim(10,20)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})
plt.show()
```

# Градиентный спуск (ГС, Gradient Descent, GD)

Градиентный спуск -- семейство методов градиентной оптимизации, наиболее популярных в задачах машинного и глубокого обучения.

### Идея

Метод используется для итеративного поиска локального (потенциально глобального) минимума (или максимума) или для улучшения результата (в терминах задачи оптимизации) любой заданной позиции.

### Формулировка и алгоритм

Пусть задана целевая функция $F(x)$, отображающая некое множество $\mathbb{X}$ на множесто действительных чисел $\mathbb{R}$. Необходимо найти минимум функции на множестве $\mathbb{X}$: $F(x) \rightarrow \min_{x \in \mathbb{X}}$ (для поиска максимума аналогично ищется минимум $-F(x)$).

Для решения задачи оптимизации методом градиентного спуска выберем произвольную начальную точку $x_{0} \in \mathbb{X}$. Будем итеративно двигаться от текущей с некоторым шагом в сторону антиградиента ($-\nabla F$).

$$
\begin{align}
x^{[0]} &= x_{0}, \\ 
x^{[i+1]} &= x^{[i]} - \lambda \nabla F(x^{[i]}),
\end{align}
$$

где $x^{[i]}$ -- некоторая точка, полученная в процессе работы алгоритма на $i$-ой итерации, $\lambda \in \mathbb{R}$ -- шаг итерации (или скорость). Шаг может быть константой или меняться во время работы алгоритма. Момент остановки алгоритма определяется по некоторому правилу, например, $|F(x^{[i+1]}) - F(x^{[i]})| \le \epsilon$.

### Интерпретация

### ГС в задачах машинного обучения

$$
\theta^{[i+1]} = \theta^{[i]} - \eta \nabla_{\theta} \mathcal{L(\theta)}
$$

 - $\theta$ -- параметры модели;
 - $\eta$ -- скорость обучения (learning rate), параметр обучения или гиппер-параметр;
 - $\nabla_{\theta} \mathcal{L}$ -- градиент по параметрам $\theta$ функции цели $\mathcal{L}$.


### Масшатбирование признаков

Градиентный спуск работает хорошо, если линии уровня функции похожи на круги, как это показано на изображении 1. В этом случае, откуда бы мы не начинали, вектор градиента будет смотреть в сторону минимума функции и будет сходиться довольно быстро.
Если же линии уровня похожи на эллипсы, как на изображении 2, то градиентный спуск будет проблемным. Направление антиградиента будет слабо совпадать с направлением в сторону минимума функции, и градиентный спуск будет делать много лишних шагов, а его сходимость будет медленной. Более того, если риск расхождения градиентного спуска.

Градиентный спуск 1-с масштабированием;2-без масштабирования

### Скорость сходимости
Пусть  𝜖  достаточно малое число, задающее желаемую точность (погрешность) при поиске оптимума. Тогда, можно доказать, что ГС для большинства выпуклых функций сойдется к точке минимума с точностью  𝜖  за линейное относительно  𝜖  число шагов.

### ГС для линейной регрессии и $\mathcal{L} = MSE$

$$
\nabla_{\theta} MSE(\theta) = \frac{2}{m} X^{T}(X \theta - y)
$$

```python
def gd_linear_regresion(X, y, eta = 0.01, n_iter = 1000):
    m = X.shape[0]
    theta = np.random.randn(2,1)*5
    theta_path = []

    for i in range(n_iter):
        grads = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta * grads

        theta_path.append(theta)
    
    return theta, np.array(theta_path)

theta, gd_theta_path = gd_linear_regresion(X,Y)
x_for_line = np.array([0,10])
pred_line = theta[1]*x_for_line + theta[0]

fig = plt.figure(figsize=(14,4))

fig.add_subplot(1,2,1)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(x_for_line, pred_line, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

fig.add_subplot(1,2,2)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(x_for_line, pred_line, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.xlim(2,4)
plt.ylim(10,20)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})
plt.show()

print(theta)
```

# Стохастический градиентный спуск (Stochastic Gradient Descent, SGD)

 - Менее подвержен "застреванию" в точках локального минимума, особенно когда функция издержек крайне нерегулярна;
 - Меняется значения параметра скачкообразно, скорее всего никогда не попадет в точку глобального минимума, а будет "прыгать" около нее;
 - Для решения этой проблемы можно производить так называемую *имитацию отжига (simulated annealing)*. То есть ввести *график обучения (learning schedule)* -- убывающая функция, которая будет определять значение шага обучения или *learning rate* на каждой итерации. 



```python
def sgd_linear_regresion(X, y, n_epochs = 5, eta = 0.01):
    m = X.shape[0]
    theta_path = []
    theta = np.random.randn(2,1)*5

    for i in range(n_epochs):
        shuffled_indices = np.random.permutation(m)
        for i in range(m):
            #random_index = np.random.randint(m)
            xi = X_b[shuffled_indices[i]: shuffled_indices[i]+1]
            yi = y[shuffled_indices[i]: shuffled_indices[i]+1]
            grads = 2/m * xi.T.dot(xi.dot(theta) - yi)
            theta = theta - eta * grads

        theta_path.append(theta)
  
    return theta, np.array(theta_path)

theta, sgd_theta_path = sgd_linear_regresion(X,Y)
x_for_line = np.array([0,10])
pred_line = theta[1]*x_for_line + theta[0]

fig = plt.figure(figsize=(14,4))

fig.add_subplot(1,2,1)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(x_for_line, pred_line, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

fig.add_subplot(1,2,2)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(x_for_line, pred_line, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.xlim(2,4)
plt.ylim(10,20)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})
plt.show()

print(theta)
```

# Мини-пакетный стохастический градиентный спуск (Mini-batched SGD)

Это крайний рассматриваемый алгоритм градиентного спуска(ГС). Понять его довольно просто, т.к. мы уже познакомились с пакетным и стохастическим ГС. На каждом шаге вместо вычисления градиентов на основе полного обучающего набора (как в пакетном ГС) или только одного образца ( как в стохастическом ГС) мини-пакетный ГС вычисляет градиентны на небольших случайных наборах образцов, которые называются **мини-пакетами**. Главное превосходство мини-пакетного ГС над стохастическим ГС в том, что вы можете получить подъем производительности от аппаратной оптимизации матричных операций, особенно когда используются графические процессоры.
В результате мини-пакетный ГС закончит блуждения чуть ближе к минимуму, чем SGD. Но с другой стороны ему может быть труднее уйти от локальных минимумов.

```python
def batched_sgd_linear_regresion(X, y, n_epochs = 10, minibatch_size = 2, eta = 0.01):
    m = X.shape[0]
    theta_path = []
    theta = np.random.randn(2,1)*5

    for i in range(n_epochs):
        shuffled_indices = np.random.permutation(m)
        for i in range(0, m, minibatch_size):
            xi = X_b[shuffled_indices]
            yi = y[shuffled_indices]
            grads = 2/m * xi.T.dot(xi.dot(theta) - yi)
            theta = theta - eta * grads

        theta_path.append(theta)
  
    return theta, np.array(theta_path)

theta, batched_sgd_theta_path = batched_sgd_linear_regresion(X,Y)
x_for_line = np.array([0,10])
pred_line = theta[1]*x_for_line + theta[0]

fig = plt.figure(figsize=(14,4))

fig.add_subplot(1,2,1)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(x_for_line, pred_line, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.ylim(-10,60)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})

fig.add_subplot(1,2,2)
plt.plot(dX, Y_gt, "g-", linewidth=1, label='Истиный закон')
plt.plot(x_for_line, pred_line, "b-", linewidth=1, label='Аппроксимация')
plt.plot(X, Y, 'r.', markersize=8)
plt.xlim(2,4)
plt.ylim(10,20)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(prop={'size': 14})
plt.show()

print(theta)
```

# Применимость градиентного спуска для выпуклых функций

### Субдифференциал и субградиент

Пусть $f(x) : E \rightarrow \mathbb{R}$ функция, определенная на множестве $E$ в евклидовом пространстве $V$, и $x_{0} \in E$. Вектор $s \in V$ называется *субградиентом функции* $f$ *в точке* $x_{0}$, если:

$$
\begin{equation}
f(x) \ge f(x_{0}) + \langle s, x - x_{0} \rangle
\tag{1}
\end{equation}
$$

для всех $x \in E$, где $\langle , \rangle$ -- некоторое скалярное произведение в пространстве $V$, далее стандартное скалярное произведение. Множество всех субградиентов в точке $x_{0}$ функции $f$ называется *субдифференциалом функции в точке* $x_{0}$ и обозначается $\delta f(x_{0})$.  Функция $\delta f(x)$, ставящая каждой точке $x \in E$ в соответствие множество $\delta f(x)$ в пространстве $V$, называется *субдифференциалом функции* $f$.

Если $\delta f(x_{0}) \neq \emptyset$, где $\emptyset$ -- пустое множество, то функция $f$ называется *субдифференцируемой в точке* $x_0$; если $f$ субдифференцируема в любой точке области определения, то $f$ называется просто *субдифференцируемой*.

Факты (без доказательства):

- Пусть $f : E \rightarrow \mathbb{R}$ — функция, определенная на выпуклом множестве $E$ в евклидовом пространстве. Если $f$ субдифференцируемая, то $f$ выпуклая.
- Пусть $f : E \rightarrow \mathbb{R}$ — функция, определенная на множестве
$E$ в евклидовом пространстве, и пусть $x_0 \in E$. Тогда точка $x_0$ является минимумом функции $f$, если и только если $0 \in \delta f(x_0)$.
- Не всякая выпуклая функция является субдифференцируемой. Тем не менее оказывается, что отсутствие субградиентов у выпуклой функции — это довольно экзотическая ситуация, которая может произойти только в граничных точках области определения. Один из основных результатов теории субдифференциалов состоит в том, что в любой внутренней точке области определения выпуклая функция гарантированно имеет субградиенты, причем субдифференциал является не просто выпуклым и замкнутым множеством, но также ограниченным.

**Вывод**: практически все выпуклые функции субдифференцируемы, и имеют минимум в точке $x_{0}$, если $0 \in \delta f(x_{0})$. Следовательно для выпуклых функция можно аналогично использовать методы градиентной оптимизации (наподобии градиентного спуска), заменяя градиент функции на любой субградиент. Так как движение не обязательно выполняется в сторону антиградиента, необходимо на каждой итерации алгоритма проверять текущую точку на минимум.

### Пример

Пусть $f(x)=|x|$, найдем субдифферециал $\delta f(0)$.

Запишем определение субградиента $(1)$ для функции $f(x)$ и найдем все субградиенты в точке $x=0$:

$$
|x| \ge |0| + \langle s, x - 0 \rangle = sx.
$$

Рассмотрим три случая:

Пусть $x > 0$, тогда $x \ge sx \Rightarrow s \le 1$.

Пусть $x = 0$, тогда $s$ -- любое.

Пусть $x < 0$,  тогда $-x \ge sx \Rightarrow s \ge -1$.

Следовательно $s \in [0;1]$ или $\delta f(0) = [0;1]$.

Также заметим, что $0 \in [0;1]$, следовательно при $x = 0$ достигается минимум $f(x) = |x|$.

# Применимость градиентного спуска для произвольных функций

### Случай 1. Градиент (производную) не удается аналитически вычислить или градиент не существует в точке

Можно воспользоваться методами вычислительной математики, например, методом конечных разностей и считать значение градиента приближенно. Для примера, возьмем функцию $f(x)=|x|$ и приближенно с точностью $\epsilon$ посчитаем производную в точке $x=0$:

$$
f'(x) \approx \frac{f(x - 2\epsilon) - 8f(x - \epsilon) + 8 f(x + \epsilon) - f(x + 2\epsilon)}{12 \epsilon}
$$

</blockquote>

```python
def f(x):
    return np.abs(x)

def f2(x):
    if x < 0:
        return 0
    elif x > 0:
        return 1
    else:
        return 0.5

def df(x, f, epsilon = 0.001):
    return (f(x-2*epsilon) - 8*f(x-epsilon) + 8*f(x+epsilon) - f(x+2*epsilon))/(12*epsilon)

print('с точностью epsilon = ', 0.001,', df(0)/dx = ', df(0, f, 0.001))
print('с точностью epsilon = ', 0.001,', df2(0)/dx = ', df(0, f2, 0.00001))
```

<blockquote>

### Случай 2. Функция не дифференцируемая

Заменяем фунцию на "похожую", то есть на некоторую приближенную функцию или апроксимацию или интерполяцию, для которой рассчитать градиент можно либо аналитически, либо численно.

Альтернативно, заменяем функцию на дифференцируемую, которая будет возрастать и убывать аналогично исходной.

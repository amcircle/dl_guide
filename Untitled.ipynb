```python
import tensorflow as tf
import numpy as np
```

```python
from glob import glob

dataset_path = './101_ObjectCategories/'

class_folders = glob(dataset_path + '*')
```

```python
class_folders
```

```python
per_class_images = {}

for c in class_folders:
    class_name = c.split('/')[-1] # название папки / класса
    per_class_images[class_name] = glob(c+'/*.jpg') # список всех путей всех изображений
```

```python
for c in per_class_images:
    per_class_images[c] = tf.data.Dataset.from_tensor_slices(per_class_images[c])
```

```python
for i,c in enumerate(per_class_images):
    per_class_images[c] = per_class_images[c].map(lambda x: (x, i))
```

```python
dataset = list(per_class_images.values())[0]
```

```python
dataset
```

```python
for d in list(per_class_images.values())[1:]:
    dataset = dataset.concatenate(d)
```

```python
len(dataset)
```

```python
class_num = len(per_class_images)
print(class_num)
```

```python
dataset = dataset.shuffle(len(dataset), reshuffle_each_iteration=False)
```

```python
h = 200
w = 200
batch_size = 8
```

```python
import tensorflow_addons as tfa
```

```python
def proc_data(x,y):
    x = tf.io.read_file(x)
    x = tf.io.decode_jpeg(x, channels=3)
    x = tf.image.resize(x, [h,w])
    x *= 1/255
    y = tf.one_hot(y, class_num)
    return x, y

def augment(x,y):
    x = tf.image.random_flip_left_right(x)
    x = tf.image.random_flip_up_down(x)
    x = x + tf.random.normal(x.shape, mean=0.0, stddev=0.15)
    rand_angle = tf.random.normal([], mean=0.0, stddev=0.2)
    x = tfa.image.rotate(x, rand_angle)
    return x,y

#dataset = dataset.map(lambda x,y: proc_data(x,y))

split_ratio = 0.8

train_split_last = int(len(dataset)*split_ratio)

train_dataset = ( dataset
                 .take(train_split_last)
                 .shuffle(train_split_last, reshuffle_each_iteration=True)
                 .map(lambda x,y: proc_data(x,y))
                 .map(lambda x,y: augment(x,y))
                 .batch(batch_size)
                 .prefetch(tf.data.AUTOTUNE)
                )
test_dataset = ( dataset
                .skip(train_split_last)
                .map(lambda x,y: proc_data(x,y))
                .batch(batch_size)
                .prefetch(tf.data.AUTOTUNE)
               )
```

```python
from matplotlib import pyplot as plt

for x,y in train_dataset.take(1):
    print(x.shape, y.shape)
    plt.imshow(x[0])
    plt.show()
    print(tf.argmax(y[0]), y[0])
    print(x[0,10:20,10:20,:])
```

```python
tf.keras.backend.clear_session()
```

```python
input_im = tf.keras.Input(shape=(h, w, 3))
x = tf.keras.layers.Flatten()(input_im)
x = tf.keras.layers.Dense(400, 
                          activation='sigmoid',
                         kernel_initializer='glorot_uniform')(x)
x = tf.keras.layers.Dense(100, 
                          activation='sigmoid', 
                          kernel_initializer='glorot_uniform')(x)
x = tf.keras.layers.Dense(class_num, 
                          activation='softmax', 
                          kernel_initializer='glorot_uniform')(x)

better_model = tf.keras.Model(input_im, x)

better_model.summary()
```

```python
better_model.compile(
    optimizer="sgd",
    loss='categorical_crossentropy',
    metrics='accuracy')

h_better_model = better_model.fit(
    train_dataset, 
    validation_data = test_dataset, 
    epochs = 20)
```

```python
def plot_history(h):
  plt.plot(h.history['accuracy'])
  plt.plot(h.history['val_accuracy'])
  plt.title('model accuracy')
  plt.ylabel('accuracy')
  plt.xlabel('epoch')
  plt.legend(['train', 'val'], loc='upper left')
  plt.show()

  plt.plot(h.history['loss'])
  plt.plot(h.history['val_loss'])
  plt.title('model loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['train', 'val'], loc='upper left')

  plt.show()
```

```python
plot_history(h_better_model)
```

```python
input_im = tf.keras.Input(shape=(h, w, 3))
x = tf.keras.layers.Conv2D(32,3, 
                          activation='relu',
                           padding='same',
                           strides=2,
                         kernel_initializer='he_uniform')(input_im)
x = tf.keras.layers.Conv2D(64,3, 
                          activation='relu',
                           padding='same',
                           strides=2,
                         kernel_initializer='he_uniform')(x)
x = tf.keras.layers.Conv2D(128,3, 
                          activation='relu',
                           padding='same',
                           strides=2,
                         kernel_initializer='he_uniform')(x)
x = tf.keras.layers.Conv2D(128,3, 
                          activation='relu',
                           padding='same',
                           strides=2,
                         kernel_initializer='he_uniform')(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(class_num, 
                          activation='softmax', 
                          kernel_initializer='glorot_uniform')(x)

model_cnn = tf.keras.Model(input_im, x)

model_cnn.summary()
```

```python
model_cnn.compile(
    optimizer="adam",
    loss='categorical_crossentropy',
    metrics='accuracy')

h_model_cnn = model_cnn.fit(
    train_dataset, 
    validation_data = test_dataset, 
    epochs = 20)
```

```python
plot_history(h_model_cnn)
```
